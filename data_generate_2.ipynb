```sage
from sage.all import Permutations
import csv

n = 7

# store the data as a list
data = []

# generate all permutations of length 7
for p in Permutations(n):
    data.append({
        "n": n,
        "permutation": list(p),
        "coxeter_length": p.length()
    })

with open("length_7.csv", "w", newline="") as file:
    writer = csv.writer(file)

    # Header
    writer.writerow(["n", "permutation", "coxeter_length"])

    # Data
    for row in data:
        writer.writerow([
            row["n"],
            row["permutation"],
            row["coxeter_length"]
        ])

print("Saved length_7.csv")
```
