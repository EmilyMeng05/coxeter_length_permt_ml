**Project purposes:** 

For this project, we aim to train a machine model to predict the Coxeter length of a given permutation. 

**Some background information:** 

Coxeter length of a permutation equals the number of inversions of a permutation 

An inversion is a pair i &lt; j with w(i) &gt; w(j), with w(k) being the place k gets sent to in the one-line notation for permutations.

For example, in S5, w = [3,5,1,4,2]. w(1) = 3, which means 1 gets sent to 3. 

The number of inversions for a permutation in a Symmetric group n is given by summing the elements of the inversion vector. In sage, the function is already defined. We can just use **number_of_inversions().** 

[link for the number_of_inversion function](https://doc.sagemath.org/html/en/reference/combinat/sage/combinat/permutation.html#sage.combinat.permutation.Permutation.number_of_inversions).

OR! use directly **.length()** which gives the Coxeter length.

<span style='font-size:large'>**Part 1: Generate the data set using Sage**</span>

<span style='font-size:medium'><span style='font-size:large'>I started with generating permutations with lengths 1 to 6 (inclusive) </span></span>


```sage
from sage.all import Permutations

# store the data as a dictionary
data = []

# generate all the permutation with lengths 1 to 6 (inclusive)
for n in range(1, 7):
    for p in Permutations(n):
        data.append({
            "n": n,
            "permutation": list(p),
            "coxeter_length": p.length()
        })
```

Part 1.2 Explore the data set


```sage
import csv

with open("coxeter_length_data.csv", "w", newline="") as file:
    writer = csv.writer(file)

    # Header
    writer.writerow(["n", "permutation", "coxeter_length"])

    # Data
    for row in data:
        writer.writerow([
            row["n"],
            row["permutation"],
            row["coxeter_length"]
        ])

print("Saved coxeter_length_data.csv")
```
